```python
#Boxplot of all the columns for statistical analysis
sns.set(style="ticks")
plt.figure(figsize=(10,10))
for j,attr in enumerate(list(df.columns.values)):
  plt.subplot(4,4,(j+1))
  sns.boxplot(y=df[attr])
  plt.grid()
  plt.tight_layout()
```

```python
#handling outliers
for j,attr in enumerate(list(df.columns.values)):
        quartile1=df[attr].quantile(0.25)
        quartile3=df[attr].quantile(0.75)
        iqr=quartile3-quartile1
        lower = quartile1-1.5*iqr
        upper = quartile3+1.5*iqr
        df = df[df[attr]< upper]
        df= df[df[attr]>lower]
```

```python
#Information about dataset after handling outliers
df.info()
```

```python
#correlation between columns
corr = df[df.columns].corr()
corr
```

```python
#heatmap of the correlation
hide = np.zeros_like(corr)
hide[np.triu_indices_from(hide)] = True
with sns.axes_style("darkgrid"):
    f, ax = plt.subplots(figsize=(13, 11))
    ax = sns.heatmap(corr,mask=hide, vmax=1, square=True,annot=True)
```

```python
#as its correlation is only -0.08
X=df.drop('Ash',axis=1)
```

```python
#as it is to be predicted
X=X.drop(['Class'], axis=1)
```

```python
#Class needs to be predicted
y=df['Class']
```

```python
from sklearn.model_selection import train_test_split
from sklearn.model_selection import KFold
from sklearn.neighbors import KNeighborsClassifier
from sklearn.naive_bayes import GaussianNB
```

```python
#Splitting of data for training and testing
X_train, X_test, y_train, y_test = train_test_split( X, y, test_size=0.30, random_state=42)
```

 > # **Applying models**

**1. KNN**

```python
model1=KNeighborsClassifier(n_neighbors=7)
```

```python
model1.fit(X_train,y_train)
```

```python
y_pred=model1.predict(X_test)
```

```python
y_pred
```

```python
from sklearn.metrics import confusion_matrix
from sklearn.metrics import accuracy_score
from sklearn.metrics import classification_report
```

```python
#Accuracy
print(accuracy_score(y_test,y_pred))
```

```python
print(classification_report(y_test,y_pred))
```

```python
matrix = confusion_matrix(y_test,y_pred)
sns.heatmap(matrix.T, square=True, annot=True, fmt='d', cbar=True,cmap="Reds")
plt.xlabel('true classes')
plt.ylabel('predicted classes')
plt.show()
```

**2. Naive Bayes**

```python
#Using Naive Bayes as our model
model=GaussianNB()
```

```python
#Fitting the training dataset in model
model.fit(X_train,y_train)
```

```python
#Prediction values of the test dataset
y_pred=model.predict(X_test)
```

```python
#Predicted Classes
y_pred
```

```python
#Test Classes
y_test
```

```python
#Accuracy
print(accuracy_score(y_test,y_pred))
```

```python
print(classification_report(y_test,y_pred))
```
